Data Filtering
process of selecting rows or record that meet a certain criteria. for instance we might be intersted in getting year and price of the oil prices

```python
df.loc[:, ["year", "Price"]]
```

Next we going to select rows 0-10 and all their columns

```python
df.loc[0:10, ["year", "Price"]]
```

Next we want rows that meet a certain criteria, for intance select rows 2021 on the month of December

```python
df.loc[(df.year ==2021) & (df.month ==12), : ]
```

if you want to check in any of the column values for a rows with null values. only rows with null are returned

```python
df.loc[df.isnull().any(axis=1), :]
```

we can also check the specific column you want checked for nulls and also calculate the number of nulls, We will use column "Vol"

```python
df.loc[df["Vol"].isnull(), :]
```

```python
df["Vol"].isnull().sum()
```
